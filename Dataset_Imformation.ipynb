```python
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
defocus_path = 'blur_dataset/defocused_blurred/0_IPHONE-SE_F.JPG'
defocus = cv2.imread(defocus_path)
defocus = cv2.cvtColor(defocus, cv2.COLOR_BGR2RGB)
motion_path = 'blur_dataset/motion_blurred/0_IPHONE-SE_M.JPG'
motion = cv2.imread(motion_path)
motion = cv2.cvtColor(motion, cv2.COLOR_BGR2RGB)
sharp_path = 'blur_dataset/sharp/0_IPHONE-SE_S.JPG'
sharp = cv2.imread(sharp_path)
sharp = cv2.cvtColor(sharp, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.imshow(defocus)

plt.subplot(1, 3, 2)
plt.imshow(motion)

plt.subplot(1, 3, 3)
plt.imshow(sharp)
```

```python
defocus_shape = defocus.shape
motion_shape = motion.shape
sharp_shape = sharp.shape
print(defocus_shape)
print(motion_shape)
print(sharp_shape)
```

```python
import os

def count_files_in_folder(folder_path):
    try:
        files = os.listdir(folder_path)
        num_files = len(files)
        return num_files
    except OSError as e:
        print(f"Error reading folder {folder_path}: {e}")
        return None

# Specify the paths of the three folders
folder1_path = "blur_dataset/defocused_blurred"
folder2_path = "blur_dataset/motion_blurred"
folder3_path = "blur_dataset/sharp"

# Count files in each folder
num_files_folder1 = count_files_in_folder(folder1_path)
num_files_folder2 = count_files_in_folder(folder2_path)
num_files_folder3 = count_files_in_folder(folder3_path)

# Print the results
print(f"Number of files in {folder1_path}: {num_files_folder1}")
print(f"Number of files in {folder2_path}: {num_files_folder2}")
print(f"Number of files in {folder3_path}: {num_files_folder3}")
```
